## Two ways to fill missing values
#### 1. Mode as Mode

```python
VIC['Display'].fillna('15.6', inplace=True)
VIC['Display Resolution'].fillna(VIC['Display Resolution'].mode()[0], inplace=True)
VIC['Display Refresh Rate'].fillna(VIC['Display Refresh Rate'].mode()[0], inplace=True)
VIC['Weight'].fillna(VIC['Weight'].mode()[0], inplace=True)
VIC['Video graphics'].fillna(VIC[''].mode()[0], inplace=True)
```

#### 2. OR Mode as Value

```python
VIC['Display'].fillna('15.6', inplace=True)
VIC['Display Resolution'].fillna('1920 x 1080', inplace=True)
VIC['Display Refresh Rate'].fillna('144 Hz', inplace=True)
VIC['Weight'].fillna('2.29 kg', inplace=True)
```

#### Extracting numeric values from text columns

Using regex to extract numbers from text columns and convert them to numeric types:
- RAM: Extract first number (GB)
- Weight: Extract first number (kg)
- Screen size: Extract decimal number (inches)
- Display Refresh Rate: Extract number (Hz)
- GPU: Extract GPU model number
- Storage: Extract storage capacity number
- CPU: Extract processor generation number (4+ digits)


```python
VIC_df['RAM'] = VIC_df['RAM'].str.extract(r'(\d+)').astype(int)
```

```python
VIC_df['Weight'] = VIC_df['Weight'].str.extract(r'(\d+)').astype(float)
```

```python
VIC_df['Screen size'] = VIC_df['Display'].str.extract(r'(\d+\.?\d+)').astype(float)
```

```python
VIC_df['Display Refresh Rate'] = VIC_df['Display Refresh Rate'].astype(str).str.extract(r'(\d+)').astype(int)
```

```python
VIC_df['GPU'] = VIC_df['Video graphics'].str.extract(r'(\d+)').astype(float)
```

```python
VIC_df['Storage'] = VIC_df['Hard drive'].str.extract(r'(\d+-?\d+)').astype(float)
```

```python
VIC_df['CPU'] = VIC_df['Processor'].str.extract(r'(\d{4,})').astype(float)
```

#### Converting Storage values from TB to GB

Converting storage capacity values: 1 TB = 1000 GB, 2 TB = 2000 GB


```python
VIC_df['Storage'].loc[VIC_df['Storage'] == 1] =  1000
VIC_df['Storage'].loc[VIC_df['Storage'] == 2] =  2000
```

#### Splitting Display Resolution into Width and Height + Setting it as float ( and price to float too)

```python
VIC_df[['Screen Width', 'Screen Height']] = VIC_df['Display Resolution'].str.extract(r'(\d{3,4})\D+(\d{3,4})')
VIC_df['Screen Width'] = VIC_df['Screen Width'].astype(float)
VIC_df['Screen Height'] = VIC_df['Screen Height'].astype(float)
VIC_df['Price'] = VIC_df['Price'].astype(float)
```

```python
VIC_df.info()
```

#### Filling missing values with median

```python
VIC_df['GPU'].fillna(VIC_df['GPU'].median(), inplace=True)
VIC_df['CPU'].fillna(VIC_df['CPU'].median(), inplace=True)
VIC_df['Storage'].fillna(VIC_df['Storage'].median(), inplace=True)
```

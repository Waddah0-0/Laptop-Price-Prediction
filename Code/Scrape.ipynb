```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

## Scrapping laptops info from different categories :
######  gamming laptops from (1 - 24) pages (Maximum Pages) 
######  personal laptops from (1 - 13) pages (Maximum Pages)
######  business laptops from (1 - 6) pages (Maximum Pages)
######  graphics and design laptops from (1) only one page

```python
# https://egyptlaptop.com/laptops-category
laptops_categories = ["gaming-laptops-catlog", "personal-laptops" ,  "business-laptops-catlog" ,"graphics-and-design-laptops"]
```

#### count the number of fetched laptops:

```python
counter=0
laptops_url_list=[]
```

#### Get all laptops titles from all pages in Gamming Laptops Category: (1 - 24) pages

```python
for i in range(24):
    url =f"https://egyptlaptop.com/laptops-category/{laptops_categories[0]}/page-{i+1}/?items_per_page=32"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "lxml")
    all_titles = soup.find_all("a", {"class":"product-title"})
    counter += len(all_titles)
  
    for title in all_titles:
        link = title.attrs['href']
        laptops_url_list.append(link)

print(f"Total laptops: {counter}")
```

#### Get all laptops titles from all pages in Personal Laptops Category: (1 - 13) pages

```python
for i in range(13):
    url =f"https://egyptlaptop.com/laptops-category/{laptops_categories[1]}/page-{i+1}/?items_per_page=32"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "lxml")
    all_titles = soup.find_all("a", {"class":"product-title"})
    counter += len(all_titles)
   
    for title in all_titles:
        link = title.attrs['href']
        laptops_url_list.append(link)

print(f"Total laptops: {counter}")
```

#### Get all laptops titles from all pages in Business Laptops Category: (1 - 6) pages

```python
for i in range(6):
    url =f"https://egyptlaptop.com/laptops-category/{laptops_categories[2]}/page-{i+1}/?items_per_page=32"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "lxml")
    all_titles = soup.find_all("a", {"class":"product-title"})
    counter += len(all_titles)
    
    for title in all_titles:
        link = title.attrs['href']
        laptops_url_list.append(link)

print(f"Total laptops: {counter}")
```

##### Define a list that takes all laptop info as dictionary

```python
all_laptops_data=[]
```

##### Get all the laptop info from the laptop page details ( labels, values)

```python
for i in range(len(laptops_url_list)):
    laptop_page_url =f"{laptops_url_list[i]}"
    laptop_page = requests.get(laptop_page_url)
    soup = BeautifulSoup(laptop_page.content, "lxml")
    
    price = soup.find("span", {"class":"ty-price-num"})
    features_list = soup.find_all("div" ,{"class":"ty-product-feature"})
    
    # laptop information dictionary
    laptop_info={}
    
    for feature in features_list: 
        # list of features labels
        labels = feature.contents[1].text.strip()
        # list of features values
        values = feature.contents[3].text.strip()
        laptop_info[labels]= values
       

    laptop_info["Price"] = int(price.text.strip().replace(",", ""))
    all_laptops_data.append(laptop_info)
```

##### Get all dictionaries lables(keys) as a list to write header row in csv file:

```python
all_keys = set()
for laptop in all_laptops_data:
    all_keys.update(laptop.keys())

fieldnames = sorted(all_keys)

print(fieldnames)
```

##### Write all laptops data in csv file

```python
with open("laptops_dataset.csv", "w", newline="", encoding="utf-8-sig") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(all_laptops_data)
```
